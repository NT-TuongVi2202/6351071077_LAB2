```python
import pandas as pd
import numpy as np

from sklearn.cluster import KMeans
from sklearn import metrics

import seaborn as sns
import plotly as py
import plotly.graph_objs as go
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv(r'c:\Users\PC\Downloads\Mall_Customers.csv')
print('Nguyễn Thị Tường Vi - 6351071077')
print(df.head())
```

```python
print('Nguyễn Thị Tường Vi - 6351071077')
print("\nThông tin dataset:")
print(df.info())

print("\nThống kê mô tả các cột số:")
print(df.describe())

print("\nKiểm tra số giá trị thiếu mỗi cột:")
print(df.isnull().sum())


```

```python

for col in df.select_dtypes(include=['float64', 'int64']).columns:
    df[col].fillna(df[col].mean(), inplace=True)

for col in df.select_dtypes(include=['object']).columns:
    df[col].fillna("Unknown", inplace=True)

print("\nSố giá trị thiếu sau khi xử lý:")
print(df.isnull().sum())
```

```python
print('Nguyễn Thị Tường Vi - 6351071077')
plt.style.use('fivethirtyeight')
plt.figure(1, figsize=(15,6))

columns_to_plot = ['Age', 'Annual Income (k$)', 'Spending Score (1-100)']

for n, col in enumerate(columns_to_plot, start=1):
    plt.subplot(1, 3, n)
    plt.subplots_adjust(hspace=0.5, wspace=0.5)
    sns.histplot(df[col], bins=20, kde=True)
    plt.title(f'Histogram of {col}')

plt.show()
```
